# Dataset: Yellow Taxi Trip Records - Março 2016

```python
import pandas as pd
import pyarrow as pa
from pathlib import Path
```

```python
df = pd.read_csv('../data/yellow_tripdata_2016-03.csv')
print(f"Total de registros: {len(df):,}")
print(f"Total de colunas: {len(df.columns)}")
df.head()
```

# 2. Data Quality & Anomalias
#### Detectar e remover registros inválidos

```python
# Converter colunas de data
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])

# Verificar valores nulos
print("=== VALORES NULOS ===")
null_counts = df.isnull().sum()
print(null_counts[null_counts > 0] if null_counts.sum() > 0 else "Nenhum valor nulo encontrado.")
```

```python
# Detectar anomalias
print("=== ANOMALIAS ===")

# Distância inválida
invalid_distance = (df['trip_distance'] <= 0) | (df['trip_distance'] > 100)
print(f"Registros com distância inválida: {invalid_distance.sum()}")

# Número de passageiros inválido
invalid_passengers = (df['passenger_count'] <= 0) | (df['passenger_count'] > 6)
print(f"Registros com número de passageiros inválido: {invalid_passengers.sum()}")

# Tarifa negativa
invalid_fare = df['fare_amount'] < 0
print(f"Registros com tarifa negativa: {invalid_fare.sum()}")

# Duração inválida (pickup depois do dropoff)
invalid_duration = df['tpep_pickup_datetime'] > df['tpep_dropoff_datetime']
print(f"Registros com duração inválida: {invalid_duration.sum()}")

# Coordenadas fora de NYC
invalid_coordinates = (
    (df['pickup_longitude'] < -74.25559) | (df['pickup_longitude'] > -73.70001) |
    (df['pickup_latitude'] < 40.49612) | (df['pickup_latitude'] > 40.91553) |
    (df['dropoff_longitude'] < -74.25559) | (df['dropoff_longitude'] > -73.70001) |
    (df['dropoff_latitude'] < 40.49612) | (df['dropoff_latitude'] > 40.91553)
)
print(f"Registros com coordenadas fora de NYC: {invalid_coordinates.sum()}")
```

```python
# Remover registros com as anomalias detectadas
mask_valid = ~(invalid_distance | invalid_passengers | invalid_fare | invalid_duration | invalid_coordinates)
removed = (~mask_valid).sum()
total_before = len(df)
df = df[mask_valid].copy()
print(f"Removidos: {removed:,} registros. {removed/total_before:.2%} do total.")
print(f"Registros restantes: {len(df):,}")
```

# 3. Transform - Colunas Calculadas

```python
import numpy as np

# Duração em minutos
df['duration_min'] = (df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']).dt.total_seconds() / 60

# Velocidade média (mph) — evita divisão por zero
df['speed_mph'] = df['trip_distance'] * 60 / df['duration_min']
df.loc[df['duration_min'] <= 0, 'speed_mph'] = np.nan

# Receita por milha
df['revenue_per_mile'] = df['total_amount'] / df['trip_distance']
df.loc[df['trip_distance'] <= 0, 'revenue_per_mile'] = np.nan

# Hora do dia, Data (sem hora) e Dia da semana
df['pickup_hour'] = df['tpep_pickup_datetime'].dt.hour
df['pickup_date'] = df['tpep_pickup_datetime'].dt.date
df['pickup_dayofweek'] = df['tpep_pickup_datetime'].dt.day_name()

# Gorjeta percentual (em %), evita divisão por zero
df['tip_pct'] = df['tip_amount'] / df['total_amount'] * 100
df.loc[df['total_amount'] <= 0, 'tip_pct'] = np.nan

# Visualizar as novas colunas
df[['duration_min','speed_mph','revenue_per_mile','pickup_hour','pickup_date','pickup_dayofweek','tip_pct']].head()
```

# 4. Agregações & Métricas
### Sumários por hora, vendor, tipo de pagamento e zona

```python

payment_map = {
    1: 'Credit card',
    2: 'Cash',
    3: 'No charge',
    4: 'Dispute'
}

summary_by_hour = (
    df.groupby('pickup_hour')
      .agg(
          trips=('pickup_hour', 'size'),
          avg_fare=('fare_amount', 'mean'),
          avg_total=('total_amount', 'mean'),
          avg_tip_pct=('tip_pct', 'mean')
      )
      .round(2)
)

summary_by_vendor = (
    df.groupby('VendorID')
      .agg(
          trips=('VendorID', 'size'),
          avg_fare=('fare_amount', 'mean'),
          avg_total=('total_amount', 'mean'),
          avg_tip_pct=('tip_pct', 'mean')
      )
      .rename(index={1: 'Vendor 1', 2: 'Vendor 2'})
      .round(2)
)

summary_by_payment = (
    df.groupby('payment_type')
      .agg(
          trips=('payment_type', 'size'),
          avg_fare=('fare_amount', 'mean'),
          avg_total=('total_amount', 'mean'),
          avg_tip_pct=('tip_pct', 'mean')
      )
      .rename(index=payment_map)
      .round(2)
)

pickup_hotspots = (
    df.groupby(['pickup_latitude', 'pickup_longitude'])
      .size()
      .sort_values(ascending=False)
      .head(10)
 )

dropoff_hotspots = (
    df.groupby(['dropoff_latitude', 'dropoff_longitude'])
      .size()
      .sort_values(ascending=False)
      .head(10)
 )

pickup_zone = (
    pickup_hotspots.reset_index(name='trips')
    .rename(columns={'pickup_latitude': 'latitude', 'pickup_longitude': 'longitude'})
    .assign(zone_type='Pickup')
)

dropoff_zone = (
    dropoff_hotspots.reset_index(name='trips')
    .rename(columns={'dropoff_latitude': 'latitude', 'dropoff_longitude': 'longitude'})
    .assign(zone_type='Dropoff')
)

summary_by_zone = (
    pd.concat([pickup_zone, dropoff_zone], ignore_index=True)
      .sort_values('trips', ascending=False)
      .head(10)
      .round(2)
)

print("=== Sumário por hora do dia ===")
print(summary_by_hour.to_string())

print("\n=== Sumário por vendor ===")
print(summary_by_vendor.to_string())

print("\n=== Sumário por tipo de pagamento ===")
print(summary_by_payment.to_string())

print("\n=== Sumário por zona (top 10 hotspots) ===")
print(summary_by_zone.to_string(index=False))
```

```python
output_dir = Path('../data/output')
output_dir.mkdir(parents=True, exist_ok=True)

parquet_path = output_dir / 'yellow_tripdata_2016-03.parquet'

rows_before, cols_before = df.shape
df.to_parquet(parquet_path, index=False)

df_parquet = pd.read_parquet(parquet_path)
rows_after, cols_after = df_parquet.shape

sem_perda = (
    rows_before == rows_after
    and cols_before == cols_after
    and list(df.columns) == list(df_parquet.columns)
)

print(f"Arquivo salvo em: {parquet_path}")
print(f"Linhas antes: {rows_before:,} | depois: {rows_after:,}")
print(f"Colunas antes: {cols_before} | depois: {cols_after}")
print(f"Verificação de perda de dados: {'NÃO houve perda' if sem_perda else 'HOUVE perda'}")
```
